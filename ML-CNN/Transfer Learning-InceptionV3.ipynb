```python
import os
import sys
import glob

# 定義 函數 獲得 訓練樣本數/驗證樣本數/分類數
def get_nb_files(directory):
  """Get number of files by searching directory recursively"""
  if not os.path.exists(directory):
    return 0
  cnt = 0
  for r, dirs, files in os.walk(directory):
    for dr in dirs:
      cnt += len(glob.glob(os.path.join(r, dr + "/*")))
  return cnt

get_train_count=get_nb_files('./train') # 训练样本个数
print("訓練樣本個數:",get_train_count)
get_classes_count= len(glob.glob('./train' + "/*"))  # 分类数
print("分類數:",get_classes_count)
get_valid_count=get_nb_files('./valid') #验证集样本个数
print("驗證集樣本個數:",get_valid_count)
```

```python
from keras.models import Model
from keras.preprocessing import image
from keras.applications.imagenet_utils import preprocess_input, decode_predictions
from keras.applications import inception_v3
```

```python
# 移除top layer
Iv3 = inception_v3.InceptionV3(weights='imagenet', include_top=False)
#Iv3.summary()
```

```python
for layer in Iv3.layers:
    layer.trainable = False
#Iv3.summary()
```

```python
# 添加全局平均池化层
from keras.layers import Dense, GlobalAveragePooling2D,Dropout
x = Iv3.output
x = GlobalAveragePooling2D()(x) # GlobalAveragePooling2D 将 MxNxC 的张量转换成 1xC 张量，C是通道数
x = Dropout(0.25)(x)
# 添加一个全連結層
x = Dense(1024, activation='relu',kernel_regularizer=regularizers.l2(0.01))(x)
x = Dropout(0.5)(x)
# 添加一个隱藏層
x = Dense(512,activation='relu',kernel_regularizer=regularizers.l2(0.01))(x)
x = Dropout(0.5)(x)
# 添加一个分类器，假设我们有3个类
predictions = Dense(3, activation='softmax')(x)
```

```python
model = Model(inputs=Iv3.input, outputs=predictions)
```

```python
# 如何決定 steps_per_epoch 的設定
step_size_train=train_batches.n//train_batches.batch_size
step_size_train
```

```python
from keras.callbacks import EarlyStopping
early_stopping_monitor = EarlyStopping(patience=5)
```

```python
# model.fit_generator
history_ft=model.fit_generator(train_batches,
                               validation_data=valid_batches,
                               steps_per_epoch=step_size_train,
                               validation_steps=4,
                               epochs=10,
                               verbose=1,
                               class_weight='auto',
                               callbacks = [early_stopping_monitor])
```

```python
# 训练的acc_loss图
def plot_training(history):
  acc = history.history['acc']
  val_acc = history.history['val_acc']
  loss = history.history['loss']
  val_loss = history.history['val_loss']
  epochs = range(len(acc))
  plt.plot(epochs, acc, 'r.')
  plt.plot(epochs, val_acc, 'r')
  plt.title('Training and validation accuracy')
  plt.figure()
  plt.plot(epochs, loss, 'r.')
  plt.plot(epochs, val_loss, 'r-')
  plt.title('Training and validation loss')
  plt.show()

plot_training(history_ft)
```

```python
# 預測函數
# 輸入：model，圖片，目標尺寸
# 輸出：預測predict
target_size=(224, 224)
def predict(model, img, target_size):  
    if img.size != target_size:
        img = img.resize(target_size)

    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    x /= 255 # Rescale by 1/255
    preds = model.predict(x)
    return preds[0]
```

```python
# 现在顶层应该训练好了，让我们开始微调 Inception V3 的卷积层。
# 我们会锁住底下的几层，然后训练其余的顶层。
# 让我们看看每一层的名字和层号，看看我们应该锁多少层呢：
for i, layer in enumerate(Iv3.layers):
   print(i, layer.name)
```

```python
# 我们选择训练最上面的两个 Inception block
# 也就是说锁住前面249层，然后放开之后的层。
for layer in model.layers[:249]:
   layer.trainable = False
for layer in model.layers[249:]:
   layer.trainable = True
```
