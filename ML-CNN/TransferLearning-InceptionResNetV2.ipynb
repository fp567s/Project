```python
import numpy as np
from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense, Flatten
from keras.optimizers import Adam
import matplotlib.pyplot as plt
from PIL import Image
from keras.models import Model
from keras.preprocessing import image
from keras.metrics import categorical_crossentropy
from keras.preprocessing.image import ImageDataGenerator
from keras.layers.normalization import BatchNormalization
from keras.applications.imagenet_utils import preprocess_input, decode_predictions
```

```python
train_path = './train'
valid_path = './valid'
test_path = './test'
```

```python
# Data Augmentation
data_generator_with_aug= ImageDataGenerator(horizontal_flip = True,
                                             width_shift_range = 0.2,
                                             height_shift_range = 0.2,
                                             rescale=1./255)
            
data_generator_no_aug= ImageDataGenerator(rescale=1./255)
```

```python
# 圖檔大小IMAGEPIXELSIZE
IPS=150
```

```python
train_batches = data_generator_with_aug.flow_from_directory(train_path, target_size=(IPS,IPS), classes=['Tiger', 'Cat','Cappuccino'], batch_size=32,shuffle=True,class_mode='categorical')
valid_batches = data_generator_no_aug.flow_from_directory(valid_path, target_size=(IPS,IPS), classes=['Tiger', 'Cat','Cappuccino'], batch_size=32,shuffle=True,class_mode='categorical')
test_batches = data_generator_no_aug.flow_from_directory(test_path, target_size=(IPS,IPS), classes=['Tiger', 'Cat','Cappuccino'], batch_size=32,shuffle=True,class_mode='categorical')
```

```python
print(train_batches.image_shape)
```

```python
from keras.applications import inception_resnet_v2
# 呼叫 model 並exclude top layer
inception_resnet_v2_model =inception_resnet_v2.InceptionResNetV2(include_top=False, weights='imagenet', input_tensor=None, input_shape=None, pooling=None, classes=1000)
```

```python
'''for layer in inception_resnet_v2_model.layers[:20]:
    layer.trainable=False
for layer in inception_resnet_v2_model.layers[20:]:
    layer.trainable=True'''
```

```python
for layer in inception_resnet_v2_model.layers:
    layer.trainable = False
```

```python
# 添加全局平均池化层
from keras.layers import Dense, GlobalAveragePooling2D,Dropout
x = inception_resnet_v2_model.output
x = GlobalAveragePooling2D()(x) # GlobalAveragePooling2D 将 MxNxC 的张量转换成 1xC 张量，C是通道数
# Drop
#x=Dropout(0.5)(x)
# 添加全连接层
x=Dense(1024,activation='relu')(x) #we add dense layers so that the model can learn more complex functions and classify for better results.
x=Dropout(0.50)(x)
#x=Dense(1024,activation='relu')(x) #dense layer 2
#x=Dropout(0.50)(x)
x=Dense(512,activation='relu')(x) #dense layer 3
x=Dropout(0.25)(x)
# 添加一个分类器，假设我们有3个类
predictions = Dense(3, activation='softmax')(x)#final layer with softmax activation
```

```python
model = Model(inputs=inception_resnet_v2_model.input, outputs=predictions)
```

```python
#optimizer='rmsprop'
model.compile(optimizer='Adam', loss='categorical_crossentropy',metrics=['accuracy']) 
#model.summary()
```

```python
#取出存檔權重
try:
    model.load_weights("SaveModel/Tiger-Cat-Cappuccino-model-base-InceptionResNetV2.h5")
    print("載入模型成功!繼續訓練模型")
except :    
    print("載入模型失敗!開始訓練一個新模型")
```

```python
step_size_train=train_batches.n//train_batches.batch_size
step_size_train
```

```python
# model.fit_generator
history_ft=model.fit_generator(train_batches,
                               validation_data=valid_batches,
                               steps_per_epoch=step_size_train,
                               epochs=5,
                               verbose=1,
                               class_weight='auto')
```

```python
# 训练的acc_loss图
def plot_training(history):
  acc = history.history['acc']
  val_acc = history.history['val_acc']
  loss = history.history['loss']
  val_loss = history.history['val_loss']
  epochs = range(len(acc))
  plt.plot(epochs, acc, 'r.')
  plt.plot(epochs, val_acc, 'r')
  plt.title('Training and validation accuracy')
  plt.figure()
  plt.plot(epochs, loss, 'r.')
  plt.plot(epochs, val_loss, 'r-')
  plt.title('Training and validation loss')
  plt.show()

plot_training(history_ft)
```

```python
img_path = './valid/Cappuccino/3196083819_b77371ddf1.jpg.jpg'
img = image.load_img(img_path, target_size=(IPS, IPS))
```

```python
# 預測函數
# 輸入：model，圖片，目標尺寸
# 輸出：預測predict
target_size=(IPS, IPS)
def predict(model, img, target_size):  
    if img.size != target_size:
        img = img.resize(target_size)

    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    x /= 255 # Rescale by 1/255
    preds = model.predict(x)
    return preds[0]
```

```python
preds = predict(model, img, target_size)
preds
```

```python
labels = ('Tiger', 'Cat','Cappuccino')
def plot_preds(image, preds,labels):
  """Displays image and the top-n predicted probabilities in a bar graph
  Args:
    image: PIL image
    preds: list of predicted labels and their probabilities
  """
  plt.imshow(image)
  plt.axis('off')
  plt.figure()
  plt.barh([0,1,2], preds, alpha=0.5)
  plt.yticks([0,1,2], labels)
  plt.xlabel('Probability')
  plt.xlim(0,1.01)
  plt.tight_layout()
  plt.show()
plot_preds(img, preds,labels)
```
