```python
import requests
from bs4 import BeautifulSoup

dataurl=input('請輸入url:')
res = requests.get(dataurl)
soup = BeautifulSoup(res.text, 'lxml')
s=soup.string
res.status_code
```

```python
import os

savedpath='IMAGE-NET/'+input('請輸入資料夾名稱')+'/'
if not os.path.isdir(savedpath):  #如果資料夾不存在就建立
    os.mkdir(savedpath)
    
ss=s.split()
print('待下載圖檔總數:',len(ss))
n=1
for sss in ss:
    print(n)
    print(sss)
    n+=1
    name=sss.split("/")[-1]
    
    try:
        if res.status_code == requests.codes.ok: # Reponse 物件有兩種屬性 status.code, text
            print("取得網頁內容成功")
            with open(savedpath+name,'wb') as f:
                res2=requests.get(str(sss))
                f.write(res2.content)          
        else:
            print("取得網頁內容失敗")

    except Exception as e:
        pass
    
    print('圖片取得結束')
    
        
```

# RESIZE IMAGE

```python
import os
from PIL import Image

srcpath = 'IMAGE-NET/Cat/' 
dstpath = 'IMAGE-NET/Cat_RESIZE/'  
if not os.path.isdir(dstpath):  #如果資料夾不存在就建立
    os.mkdir(dstpath)

n=1
for fname in os.listdir(srcpath):    
    print(fname,"--正在處理中--")
    img = Image.open(srcpath + fname)
    crpim = img.resize((64,64))
    crpim.save(dstpath + fname)
    print(fname,"--存檔完畢--")
    n+=1
print("--RESIZE",n,"張","作業完畢--")
```
